## UMAP embeddings
Our goal is to use topology to see if our dataset has a "shape" associated to it.

After we find the shape, we will see if we can describe the shape in any meaningful way.

The UMAP algorithm is a topological algorithm that does both dimensional reduction and identifies clusters based on a topological notion of distance.

First, preliminary data manipulation and importing:

```python
from denseclus import DenseClus
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from seaborn import set_style
set_style("whitegrid")

from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
```

```python
# Reading in data

data = pd.read_csv('train.csv')
```

```python
# Add two categorical columns: wilderness type and soil types

wild = data.columns[11:15]
soil = data.columns[15:-1]

i=1
for x in wild:
  data[x] = data[x].apply(lambda x: x*i)
  i+=1

i=1
for x in soil:
  data[x] = data[x].apply(lambda x: x*i)
  i+=1

data['Wild'] = data[wild].sum(axis=1)
data['Soil'] = data[soil].sum(axis=1)

# Marks Wild and Soil as categorical variables for umap.
data = data.astype({'Wild': 'object'})
data = data.astype({'Soil': 'object'})

# dropping redundant soil1-40 and wild1-4 columns
data = data.drop(columns=soil)
data = data.drop(columns=wild)
data = data.drop(columns=['Id'])
# data.insert(0, "Type", data["Cover_Type"])
# data = data.astype({'Type': 'object'})
data1 = data.copy()
data = data.drop(columns=['Cover_Type'])
```

```python
data
```

## Start of Topological Data Analysis (TDA):

```python
# Initializing the model.

data_clf = DenseClus(random_state=314,
                     min_cluster_size=200,
                     n_components=3,
                     umap_combine_method='intersection_union_mapper')
```

```python
# Fitting the model. This takes awhile! 

data_clf.fit(data)
```

```python
# Says that each data point is in cluster X. Cluster -1 is considered outliers.

data['cluster_id'] = data_clf.score()
```

```python
# add back in the cover_type column

data1 = pd.read_csv('train.csv')
data['Type'] = data1['Cover_Type']
```

```python
data
```

```python
# Dictionary for our clusters

data_cluster_dict = {id: data[data['cluster_id']==id] for id in sorted(data['cluster_id'].unique())}
```

### What's actually in our clusters?

```python
# How many of each cover type are in each cluster?

for x in data_cluster_dict:
  print('Cluster=',x)
  print(data_cluster_dict[x]['Type'].value_counts())
```

### Plotting so we can visualize these clusters.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from IPython.display import Image
from matplotlib.pyplot import rcParams
```

```python
# Initialize Matplotlib styling parameters
plt.style.use('seaborn')
rcParams.update({
    'font.size': 12,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'axes.labelsize': 16,
    'axes.titlesize': 18,
    'legend.fontsize': 12

})
```

```python
# Custom color sequence for clustering (helps assign outliers to grey color)
custom_color_seq = ['#777777']
custom_color_seq.extend(px.colors.qualitative.Plotly)
```

```python
# makes groupings, embeddings

data_umap_embedding = data_clf.mapper_.embedding_
data_umap_df = pd.DataFrame(data_umap_embedding, columns=['x','y','z']) #dataframe
data_umap_df['cluster_id'] = data_clf.score()
data_groups = data_umap_df.groupby('cluster_id')
```

```python
#plots figure

fig = px.scatter_3d(
    data_umap_df.sort_values(by='cluster_id'),
    x='x',
    y='y',
    z='z',
    color=data['cluster_id'].sort_values().astype(object),
    color_discrete_sequence=custom_color_seq,
    title='Cover_Types UMAP Embedding')
fig.update_traces(
    marker = dict(size=3)
)
fig.update_layout(
    autosize=False,
    width=900,
    height=900,
    legend_title='cluster_id'
)
fig.layout.coloraxis.colorbar.title = 'dist_3'
fig.show()
```

Now we want to see if the x-axis relates to any of our features/combination of features.

```python
# Do linear regression on x
data['x'] = data_umap_df['x']
data['y'] = data_umap_df['y']
data['z'] = data_umap_df['z']
```

```python
# Pearson coefficients

print('Pearson coefficients on x:')
print('')
data.corr(method = 'pearson')['x']
```

```python
print('Pearson coefficients on y:')
print('')
data.corr(method = 'pearson')['y']
```

```python
print('Pearson coefficients on z:')
print('')
data.corr(method = 'pearson')['z']
```

```python
from sklearn.linear_model import LinearRegression
```

### Seems to be a strong linear relationship between the x-axis of the UMAP embedding and elevation. Let's plot this.

```python
regression = LinearRegression(copy_X = True)
regression = LinearRegression().fit(data['Elevation'].values.reshape(-1,1), data['x'].values.reshape(-1,1))
```

```python
# Plot the regression.

fig, ax = plt.subplots(1,1)

fig.set_figwidth(12)
fig.set_figheight(12)

ax.scatter(x=data['Elevation'], y=data['x'])

plt.plot(np.linspace(1900,3500,1000),
         regression.coef_[0][0]*np.linspace(1900,3500,1000) + regression.intercept_[0],
         'r--',
         label = "f(x)")

ax.set_xlabel('Elevation')
ax.set_ylabel('UMAP X-Axis')

top_x = data['Elevation'].min()
top_y = data['x'].max()


ax.set_title('Axis-Feature Correlation')
```

In conclusion: there's a strong linear correlation between an observation's elevation and its x-coordinate in the UMAP embedding. Because clusters are most spread along the x-axis, it seems like elevation is the most essential feature!
